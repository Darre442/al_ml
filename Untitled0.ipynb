```python
import matplotlib.pyplot as plt
marks=[95,45,67,34,78]
subjects=["english","maths","chemistry","biology","data structures"]
plt.bar(marks,subjects)
plt.title("mark of each subject")
plt.xlabel("subject")
plt.ylabel("marks")
plt.show()


```

```python
import matplotlib.pyplot as plt
marks=[95,45,67,34,78]
subjects=["english","maths","chemistry","biology","data structures"]
plt.pie(marks,
         labels=subjects)
plt.title("mark of each subject")
plt.xlabel("subject")
plt.ylabel("marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
marks=[95,45,67,34,78]
subjects=["english","maths","chemistry","biology","data structures"]
plt.pie(marks,
        labels=subjects,
        autopct='%1.1f%%')
plt.title("mark of each subject")
plt.xlabel("subject")
plt.ylabel("marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
sales=[100,120,130,160,145,170,146]
plt.scatter(days,sales)
plt.title("daliy sales")
plt.xlabel("label")
plt.ylabel("label")
plt.show()
```
